```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])
y_or = np.array([0, 1, 1, 1])

perceptron_and = Perceptron(max_iter=1000, tol=1e-3)
perceptron_and.fit(X, y_and)
predictions_and = perceptron_and.predict(X)
print("AND Function Predictions:", predictions_and)

perceptron_or = Perceptron(max_iter=1000, tol=1e-3)
perceptron_or.fit(X, y_or)
predictions_or = perceptron_or.predict(X)
print("OR Function Predictions:", predictions_or)

def plot_decision_boundary(perceptron, X, y, title):
    xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 100), np.linspace(-0.5, 1.5, 100))
    Z = perceptron.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=100)
    plt.title(title)
    plt.xlabel("Input 1")
    plt.ylabel("Input 2")
    plt.show()

plot_decision_boundary(perceptron_and, X, y_and, "AND Function Decision Boundary")
plot_decision_boundary(perceptron_or, X, y_or, "OR Function Decision Boundary")
```
